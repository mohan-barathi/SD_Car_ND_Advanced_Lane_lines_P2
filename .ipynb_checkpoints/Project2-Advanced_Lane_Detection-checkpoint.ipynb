## Utility methods:

```python
def display_side_by_side(image1, image1_name, image2, image2_name,_figsize=(8, 3)):
    """
    Display given 2 images side by side.
    """
    f, (ax1, ax2) = plt.subplots(1, 2, figsize=_figsize)
    f.tight_layout()
    ax1.imshow(image1)
    ax1.set_title(image1_name, fontsize=10)
    ax2.imshow(image2)
    ax2.set_title(image2_name, fontsize=10)
    plt.subplots_adjust(left=0., right=1, top=0.9, bottom=0.)
    
def bitwise_and_binary_images(binary_image1, binary_image2):
    """
    When 2 Binary images are given, 'and' operation is done over them.
    """
    result_binary_image = np.zeros_like(binary_image1)
    result_binary_image[(binary_image1 == 1) & (binary_image2 == 1)] = 1
    return result_binary_image

def bitwise_or_binary_images(binary_image1, binary_image2):
    """
    When 2 Binary images are given, 'or' operation is done over them.
    """
    result_binary_image = np.zeros_like(binary_image1)
    result_binary_image[(binary_image1 == 1) | (binary_image2 == 1)] = 1
    return result_binary_image
```

## Camera Calibration
#### The calibration of camera can be done using the chess board images provided in `camera_cal` folder
* As the camera calibration is a one-time process, and it provides the camera matrix for undistortion, we can have a singlton class, that handles calibration and undistortion

```python
class Camera_operations:
    """
    This class holds all the methods dealing with the camera calibration and undistortion
    All the methods are called during initialisation, except undistort_image methods.
    """
    def __init__(self):
        self.object_points_superset = []     # contains ideal chess board points
        self.image_points_superset = []      # contains the chess board points of given images
        self.eligible_images = 0             # No. of images eligible for camera calibration
        self.chessboard_images_list = []     # holds all the chess board images, for calibration
        self.mtx = []                        # Camera matrix
        self.dist = None                     # Distortion Co-efficients
    
    def read_all_images(self, root_folder):
        try:
            self.chessboard_images_list = os.listdir(root_folder)
            for i in range(len(self.chessboard_images_list)):
                self.chessboard_images_list[i] = cv2.imread(root_folder + self.chessboard_images_list[i])
        except:
            raise Exception("Error in reading the images..!")
        
    
    def calibrate_camera(self, nx_columns, ny_rows):
        try:
            for i in range(len(self.chessboard_images_list)):
                objp = np.zeros((ny_rows*nx_columns,3), np.float32)
                objp[:,:2] = np.mgrid[0:nx_columns,0:ny_rows].T.reshape(-1,2)
                gray_chess_board = cv2.cvtColor(self.chessboard_images_list[i], cv2.COLOR_BGR2GRAY)
                ret, corners = cv2.findChessboardCorners(gray_chess_board, (nx_columns, ny_rows), None)
                if ret == True:
                    if __DEBUG__:
                        # Draw and display the corners
                        cv2.drawChessboardCorners(self.chessboard_images_list[i], (nx_columns, ny_rows), corners, ret)
                        plt.imshow(self.chessboard_images_list[i])
                    self.eligible_images = self.eligible_images + 1
                    self.object_points_superset.append(objp)
                    self.image_points_superset.append(corners)

            print(self.eligible_images, " images were eligible for calibration")
            ret, self.mtx, self.dist, rvecs, tvecs = cv2.calibrateCamera(self.object_points_superset, self.image_points_superset, 
                                                               gray_chess_board.shape[::-1], None, None)
            return True
        except:
            raise Exception("Error in Calibrating the camera")
        
    def undistort_image(self, distorted_image):
        return cv2.undistort(distorted_image, self.mtx, self.dist, None, self.mtx)
        
```

## Sobel gradient masking and HLS masking
#### The sobel gradient masking and HLS masking is done using the following api's.
#### The filtering methods
* x-absolute-sobel, y-absolute-sobel, combined-magnitude-sobel, angular (or) directional sobel
* s-channel, l-channel, h-channel

```python
def abs_sobel_thresh(img, orient='x', sobel_kernel=3, sobel_thresh=(0,255)):
    """
    This filter for both sobel-x and sobel-y.
    """
    temp_img = img.copy()
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    if orient == 'x':
        abs_sobel = np.absolute(cv2.Sobel(gray, cv2.CV_64F, 1, 0))
    if orient == 'y':
        abs_sobel = np.absolute(cv2.Sobel(gray, cv2.CV_64F, 0, 1))
    scaled_sobel = np.uint8(255*abs_sobel/np.max(abs_sobel))
    binary_output = np.zeros_like(scaled_sobel)
    binary_output[(scaled_sobel >= sobel_thresh[0]) & (scaled_sobel <= sobel_thresh[1])] = 1
    return binary_output


def mag_thresh(img, sobel_kernel=3, mag_thresh=(0, 255)):    
    """
    This filter uses the combined magnitude of sobel-x and sobel-y for filtering.
    """
    gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    sobelx = cv2.Sobel(gray_image, cv2.CV_64F, 1, 0, ksize=sobel_kernel)
    sobely = cv2.Sobel(gray_image, cv2.CV_64F, 0, 1, ksize=sobel_kernel)
    gradmag = np.sqrt(sobely**2 + sobelx**2)
    scaled_gradmag = np.uint8(255*gradmag/np.max(gradmag))
    binary_output = np.zeros_like(gradmag)
    binary_output[(scaled_gradmag >= mag_thresh[0]) & (scaled_gradmag <= mag_thresh[1])] = 1
    return binary_output

def dir_threshold(img, sobel_kernel=3, thresh=(0, np.pi/2)):
    """
    This filter is for directional (or) angular sobel.
    """
    gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    sobelx = cv2.Sobel(gray_image, cv2.CV_64F, 1, 0, ksize=sobel_kernel)
    sobely = cv2.Sobel(gray_image, cv2.CV_64F, 0, 1, ksize=sobel_kernel)
    abs_sobelx = np.absolute(sobelx)
    abs_sobely = np.absolute(sobely)
    grad_direction = np.arctan2(abs_sobely, abs_sobelx)
    binary_output = np.zeros_like(grad_direction)
    binary_output[(grad_direction >= thresh[0]) & (grad_direction <= thresh[1])] = 1
    return binary_output

def combined_threshold(image, sobel_kernel=15, threshold_array={'x':(20,100), 'y':(20,100), 'mag':(30,100), 'dir':(0.7, 1.3)}):
    """
    This combined the above sobel filters.
    However, this is only for experimenting, and not used in the project.
    """
    gradx = abs_sobel_thresh(image, 'x', sobel_kernel, sobel_thresh=threshold_array['x'])
    grady = abs_sobel_thresh(image, 'y', sobel_kernel, sobel_thresh=threshold_array['y'])
    mag_binary = mag_thresh(image, sobel_kernel, mag_thresh=threshold_array['mag'])
    dir_binary = dir_threshold(image, sobel_kernel, thresh=threshold_array['dir'])

    combined_binary_image = np.zeros_like(dir_binary)
    combined_binary_image[((gradx == 1) & (grady == 1)) | ((mag_binary == 1) & (dir_binary == 1))] = 1
    return combined_binary_image

def s_channel_threshold(image, s_thresh=(0,255)):
    """
    The Filter for S-channel, in a HLS channel image.
    """
    hls_image = cv2.cvtColor(image,cv2.COLOR_BGR2HLS)
    s_channel = hls_image[:,:,2]
    s_binary_image = np.zeros_like(s_channel)
    s_binary_image[(s_channel > s_thresh[0]) & (s_channel <= s_thresh[1])] = 1
    return s_binary_image

def l_channel_threshold(image, l_thresh=(0,255)):
    """
    The Filter for L-channel, in a HLS channel image.
    """    
    hls_image = cv2.cvtColor(image,cv2.COLOR_BGR2HLS)
    l_channel = hls_image[:,:,1]
    l_binary_image = np.zeros_like(l_channel)
    l_binary_image[(l_channel > l_thresh[0]) & (l_channel <= l_thresh[1])] = 1
    return l_binary_image

def h_channel_threshold(image, h_thresh=(0,255)):
    """
    The Filter for H-channel, in a HLS channel image.
    """  
    hls_image = cv2.cvtColor(image,cv2.COLOR_BGR2HLS)
    h_channel = hls_image[:,:,0]
    h_binary_image = np.zeros_like(h_channel)
    h_binary_image[(h_channel > h_thresh[0]) & (h_channel <= h_thresh[1])] = 1
    return h_binary_image
```

## Conclusion of above observation:
* The `h-channel` is of no use in finding the lane
* The `s-channel` works really good in identifying the lane lines
* The `angular` (a.k.a `direction`) binary finds the lane in road with darker shade, but also have too much noise
    - Hence, this can be combined with `magnitude binary` (using `bitwise-and`)
* The `x sobel` and `y sobel` works good with road with darker shade and finding white lanes, so do the magnitude binary
    * Hence, the `magnitude binary` can be used alongside `s-channel` (`bitwise-or`) 
* The `l-channel` works good in road with darker shade, but when the road shade changes, no lanes are detected (pixels of enitre road is activated)
    * Hence, this can be used as a **final filter**, after the `magnitude`, `direction` and `s-channel` filtering

## Based on the above observations, the following `lane_filter` api is deduced:

```python
def lane_filter(image):
    """
    Containes the concluded final combination of filters.
    """
    threshold_array={'mag':(50,180), 'dir':(0.8, 1), 's':(170,255), 'l':(100,250)}
    mag_binary = mag_thresh(image, sobel_kernel=3, mag_thresh=threshold_array['mag'])
    dir_binary = dir_threshold(image, sobel_kernel=15, thresh=threshold_array['dir'])
    s_channel_image = s_channel_threshold(image, threshold_array['s'])
    l_channel_image = l_channel_threshold(image, threshold_array['l'])
    s_or_mag_image = bitwise_or_binary_images(s_channel_image, mag_binary)
    mag_and_dir_image = bitwise_and_binary_images(mag_binary, dir_binary)
    mag_or_dir_s_image = bitwise_or_binary_images(s_or_mag_image, mag_and_dir_image)
    mag_dir_s_l_image = bitwise_and_binary_images(mag_or_dir_s_image, l_channel_image)
    return mag_dir_s_l_image
```

## Perspective Transform:
#### The perspective change requires Perspective transform matrix M, and inverse perspective transform matrix Minv to change it back to original perspective.
* We shall have a singleton class here, that produces the 2 matrices, as this will stay the same for entire process
* We shall consider the test image straight_lines2.jpg for computing these matrices

```python
class Singleton_Perspective_Transform:
    """
    A Singleton class that produces perspecptive transform matrix
    and inverse perspective transform matrix.
    """
    def __init__(self):
        self._src_points = []
        self._dst_points = []
    
    def set_destination_points(self, dst_points):
        self._dst_points = dst_points
        
    def set_source_points(self, src_points):
        self._src_points = src_points
    
    def find_matrix(self):
        self.M = cv2.getPerspectiveTransform(self._src_points,self._dst_points)
        self.Minv = cv2.getPerspectiveTransform(self._dst_points,self._src_points)
        
    def warp_image(self, image):
        return cv2.warpPerspective(image, self.M, (image.shape[1], image.shape[0]), flags=cv2.INTER_LINEAR)
    
    def unwarp_image(self, image):
        return cv2.warpPerspective(image, self.Minv, (image.shape[1], image.shape[0]), flags=cv2.INTER_LINEAR)
```

### Instantiation of the `Singleton_Perspective_Transform` class

```python
# Instantiation of the Perspective Transform object
perspective_transform = Singleton_Perspective_Transform()
perspective_transform.set_source_points(np.float32([(200,720),(1115,720),(700,456),(582,456)]))  #manually obtained from image
perspective_transform.set_destination_points(np.float32([(200,720),(1115,720),(1115,0),(190,0)]))
perspective_transform.find_matrix()

# Wrapper api's for warping and unwarping, to make sure that the class stays singleton.
def warp_image(image):
    return perspective_transform.warp_image(image)
def unwarp_image(image):
    return perspective_transform.unwarp_image(image)
```

## Histogram and Sliding window search

#### To deduce the polynomial that fits the lane lines in the frames, **we should first filter out the pixels that belong to lane lines**. 
* Here we use histogram and sliding window search for that, as shown below
* Once a pair of polynomials for the lanes are found, the lanes in next frames can be deduced by a targeted search around that polynomial 

```python
class Frame_info:
    """
    This class holds the information of a frame,
    which can be used for any deductions.
    For now, only the center_of_lane is used to find whether
    a lane detection aligns to that of lane in previous frame.
    """
    def __init__(self):
        leftx = []
        lefty = []
        rightx = []
        righty = []
        left_fit_coeff = []
        right_fit_coeff = []
        left_real_curvature = 0.0
        right_real_curvature = 0.0
        car_offset_distance = 0.0
        car_offset_side = None
        center_of_lane = 0
        
```

```python
class Polynomial_lanes:
    """
    All the operations regarding the lane detection, 
    curvature and offset calculation and plotting it back on the original
    image is handled by this class.
    
    This also keeps track of last 5 frames processed, to deduce whether a 
    lane detection have sanity.
    """
    
    def __init__(self):
        self.lane_info_queue = []           # to keep track of last 5 frame-info
        self.nwindows = 9                   # no. of windows for sliding window search
        self.margin = 100                   # Margin for both sliding window and quick search
        self.minpix = 50                    # Minimum pixels for qualifying the sliding window to shift
        self.temp_image = None              # temp image for debugging
        self.failure_tolerance = 5          # Tolerance for insane lane detection
        self.failed_frames_count = 0        # No. of frames where quick search failed
        self.sliding_window_attempts = 0    # to keep track of sliding window attempts - video processing 
        self.search_around_polynomial = 0   # Same for quick search.
    
    def push_lane_info(self, lane_info):
        if len(self.lane_info_queue) >= 10:
            self.pop_lane_info()
        self.lane_info_queue.append(lane_info)
        
    def pop_lane_info(self):
        self.lane_info_queue.remove(self.lane_info_queue[0])
    
    def find_histogram(self, warped_binary_image):
        histogram = np.sum(warped_binary_image[int(warped_binary_image.shape[0]/2):,:], axis=0)
        return histogram
    
    def sliding_window_search(self, warped_binary_image):
        """
        Searches for activated pixels, and searches for pixels
        that belong to lane, by incrementing and sliding the window position.
        """
        #print ("Sliding window search")
        self.sliding_window_attempts = self.sliding_window_attempts +1
        histogram = self.find_histogram(warped_binary_image)
        midpoint = np.int(histogram.shape[0]/2)
        leftx_base = np.argmax(histogram[:midpoint])
        rightx_base = np.argmax(histogram[midpoint:]) + midpoint
        window_height = np.int(warped_binary_image.shape[0]/self.nwindows)
        nonzero = warped_binary_image.nonzero()
        nonzeroy = np.array(nonzero[0])
        nonzerox = np.array(nonzero[1])
        leftx_current = leftx_base
        rightx_current = rightx_base
        left_lane_inds = []
        right_lane_inds = []
        
        if __DEBUG__:
            # To visualize the sliding - window
            self.temp_image = np.dstack((warped_binary_image, warped_binary_image, warped_binary_image))
        
        for window in range(self.nwindows):
            # window boundaries in x and y (and right and left)
            win_y_low = warped_binary_image.shape[0] - (window+1)*window_height
            win_y_high = warped_binary_image.shape[0] - window*window_height
            win_xleft_low = leftx_current - self.margin
            win_xleft_high = leftx_current + self.margin
            win_xright_low = rightx_current - self.margin
            win_xright_high = rightx_current + self.margin

            if __DEBUG__:
                # windows on the visualization image
                cv2.rectangle(self.temp_image,(win_xleft_low,win_y_low),
                (win_xleft_high,win_y_high),(0,255,0), 2) 
                cv2.rectangle(self.temp_image,(win_xright_low,win_y_low),
                (win_xright_high,win_y_high),(0,255,0), 2) 

            # Identifying the nonzero pixels in x and y within the window
            good_left_inds = ((nonzeroy >= win_y_low) & (nonzeroy < win_y_high) & 
            (nonzerox >= win_xleft_low) &  (nonzerox < win_xleft_high)).nonzero()[0]
            good_right_inds = ((nonzeroy >= win_y_low) & (nonzeroy < win_y_high) & 
            (nonzerox >= win_xright_low) &  (nonzerox < win_xright_high)).nonzero()[0]

            # Append these indices to the lists
            left_lane_inds.append(good_left_inds)
            right_lane_inds.append(good_right_inds)

            # If pixels found is more than minpix pixels, recenter next window on their mean position
            if len(good_left_inds) > self.minpix:
                leftx_current = np.int(np.mean(nonzerox[good_left_inds]))
            if len(good_right_inds) > self.minpix:        
                rightx_current = np.int(np.mean(nonzerox[good_right_inds]))

        # Concatenate the arrays of indices (was a list of lists of pixels)
        try:
            left_lane_inds = np.concatenate(left_lane_inds)
            right_lane_inds = np.concatenate(right_lane_inds)
        except ValueError:
            # Avoids an error if the above is not implemented fully
            pass

        # Extract left and right line pixel positions
        leftx = nonzerox[left_lane_inds]
        lefty = nonzeroy[left_lane_inds] 
        rightx = nonzerox[right_lane_inds]
        righty = nonzeroy[right_lane_inds]
        
        if __DEBUG__:
            # Colors in the left and right lane regions
            self.temp_image[lefty, leftx] = [255, 0, 0]
            self.temp_image[righty, rightx] = [0, 0, 255]
        
        return leftx, lefty, rightx, righty


    def fit_polynomial(self, leftx, lefty, rightx, righty):
        """
        For a given pair of set of x and y co-ordinates,
        the 2nd order polynomial co-efficients are deduced.
        """
        # second order polynomial to each using `np.polyfit`
        left_fit_poly_coeff = np.polyfit(lefty, leftx, 2)
        right_fit_poly_coeff = np.polyfit(righty, rightx, 2)
        if __DEBUG__:
            print("Left Lane  :", left_fit_poly_coeff[0],"y2 +  ",left_fit_poly_coeff[1],"y +  ",left_fit_poly_coeff[2] )
            print("Right Lane :", right_fit_poly_coeff[0],"y2 +  ",right_fit_poly_coeff[1],"y +  ",right_fit_poly_coeff[2] )
        return left_fit_poly_coeff, right_fit_poly_coeff
    
    def search_around_poly(self, warped_binary_image, left_fit_coeff, right_fit_coeff):
        """
        This is faster than the sliding window search, as this filters all the activated
        pixels using their distance from given 2nd order polynomial curve equation.
        """
        #print ("Searching around a polynomial - (Quick search)")
        self.search_around_polynomial = self.search_around_polynomial + 1
        # choosing the activated pixels
        nonzero = warped_binary_image.nonzero()
        nonzeroy = np.array(nonzero[0])
        nonzerox = np.array(nonzero[1])

        left_lane_inds = ((nonzerox > (left_fit_coeff[0]*(nonzeroy**2) + left_fit_coeff[1]*nonzeroy + 
                        left_fit_coeff[2] - self.margin)) & (nonzerox < (left_fit_coeff[0]*(nonzeroy**2) + 
                        left_fit_coeff[1]*nonzeroy + left_fit_coeff[2] + self.margin)))
        right_lane_inds = ((nonzerox > (right_fit_coeff[0]*(nonzeroy**2) + right_fit_coeff[1]*nonzeroy + 
                        right_fit_coeff[2] - self.margin)) & (nonzerox < (right_fit_coeff[0]*(nonzeroy**2) + 
                        right_fit_coeff[1]*nonzeroy + right_fit_coeff[2] + self.margin)))

        leftx = nonzerox[left_lane_inds]
        lefty = nonzeroy[left_lane_inds] 
        rightx = nonzerox[right_lane_inds]
        righty = nonzeroy[right_lane_inds]
        
        if __DEBUG__:
            self.temp_image = np.dstack((warped_binary_image, warped_binary_image, warped_binary_image))
            # Colors in the left and right lane regions
            self.temp_image[lefty, leftx] = [255, 0, 0]
            self.temp_image[righty, rightx] = [0, 0, 255]
        
        return leftx, lefty, rightx, righty

    def measure_curvature_real(self, left_fit_coeff, right_fit_coeff):
        """
        Define conversions in x and y from pixels space to meters
        """
        # These are arbitrary values mentioned in udacity classroom
        
        ############################################
        # In reality, this value depends upon the  #
        # co-ordinates that we choose for warping  #
        ############################################
        
        ym_per_pix = 30/720                  # meters per pixel in y dimension
        xm_per_pix = 3.7/700                 # meters per pixel in x dimension
        ploty = np.linspace(0, 719, num=720) # to cover same y-range as image

        # radius of curvature corresponding to the bottom of the image
        y_eval = np.max(ploty)

        # Calculation of R_curve (radius of curvature)
        left_curve_rad = ((1 + (2*left_fit_coeff[0]*y_eval*ym_per_pix + left_fit_coeff[1])**2)**1.5) / np.absolute(2*left_fit_coeff[0])
        right_curve_rad = ((1 + (2*right_fit_coeff[0]*y_eval*ym_per_pix + right_fit_coeff[1])**2)**1.5) / np.absolute(2*right_fit_coeff[0])

        return left_curve_rad, right_curve_rad
    
    
    def find_offset(self, left_fit_coeff, right_fit_coeff):
        """
        Finds the offset position of car, from center of the lane.
        """
        left_base = (left_fit_coeff[0]*(719**2) + left_fit_coeff[1]*719 + left_fit_coeff[2])
        right_base = (right_fit_coeff[0]*(719**2) + right_fit_coeff[1]*719 + right_fit_coeff[2])
        center_of_lane = int((int(left_base) + int(right_base))/2)
        if center_of_lane > 640:
            side = "left"
            distance = ((center_of_lane-640)*(3.7/700))
        else:
            side = "right"
            distance = ((640-center_of_lane)*(3.7/700))
        return distance,side, center_of_lane
    
    
    def plot_on_real_frame(self, undist, warped_binary_image, left_fit_coeff, right_fit_coeff):
        """
        This api plots the lane back on the undistorted unwarped color image.
        """
        # Create an image to draw the lines on
        warp_zero = np.zeros_like(warped_binary_image).astype(np.uint8)
        color_warp = np.dstack((warp_zero, warp_zero, warp_zero))

        # Recast the x and y points into usable format for cv2.fillPoly()
        ploty = np.linspace(0, 719, num=720)# to cover same y-range as image
        left_fitx = left_fit_coeff[0]*ploty**2 + left_fit_coeff[1]*ploty + left_fit_coeff[2]
        right_fitx = right_fit_coeff[0]*ploty**2 + right_fit_coeff[1]*ploty + right_fit_coeff[2]
        pts_left = np.array([np.transpose(np.vstack([left_fitx, ploty]))])
        pts_right = np.array([np.flipud(np.transpose(np.vstack([right_fitx, ploty])))])
        pts = np.hstack((pts_left, pts_right))
        
        # lane onto the warped blank image
        cv2.fillPoly(color_warp, np.array([pts],dtype = np.int32), [0,255, 0])

        # unwarp the image with shade
        newwarp = unwarp_image(color_warp)
        
        # Combine the result with the original image
        real_lane = cv2.addWeighted(undist, 1, newwarp, 0.3, 0)
        return real_lane
    
    
    def check_sanity(self, frame_info):
        """
        checks the sanity of detected lanes, comparing the :
        1. Center of lane
        2. Curvature of lane
        """
        last_frame = self.lane_info_queue[len(self.lane_info_queue)-1]
        if not (np.absolute(last_frame.center_of_lane - frame_info.center_of_lane) < 50):
            return False
        
        average_curvature = 0
        
        for i in self.lane_info_queue:
            average_curvature = average_curvature + ((i.left_real_curvature + i.right_real_curvature)/2)
        average_curvature = int(average_curvature / len(self.lane_info_queue))
        current_curvature = int((frame_info.left_real_curvature + frame_info.right_real_curvature)/2)
        
        if not (np.absolute(average_curvature - current_curvature) < 2500):
            return False
        
        return True
        
    
    def process_frame(self, frame_image):
        """
        This is the entry point for each and every frames of the video. Here :
        1. The frames are undistorted
        2. filtered for lane
        3. filtered binary frames are warped
        4. lane detection is done on the warped binary (Sliding window or quick search)
        5. a polynomial is deduced from the pixels that belong to lane
        6. sanity is checked
        7. The lane is plotted back on the image frame.
        """
        undistorted_frame = singleton_camera_object.undistort_image(frame_image)
        binary_frame = lane_filter(undistorted_frame)
        warped_frame = warp_image(binary_frame)
        
        vfi = Frame_info()   # Object that holds all the video-frame-info
        
        # Check the history of previous frames
        # if less than 5 frame info in the queue, sliding window search is made
        
        if len(self.lane_info_queue) < 10:
            #print ("Lane info queue size :", len(self.lane_info_queue))
            vfi.leftx, vfi.lefty, vfi.rightx, vfi.righty = self.sliding_window_search(warped_frame)
            vfi.left_fit_coeff, vfi.right_fit_coeff = self.fit_polynomial(vfi.leftx, vfi.lefty, vfi.rightx, vfi.righty)
            vfi.left_real_curvature, vfi.right_real_curvature = self.measure_curvature_real(vfi.left_fit_coeff, 
                                                                                            vfi.right_fit_coeff)
            vfi.car_offset_distance, vfi.car_offset_side, vfi.center_of_lane = self.find_offset(vfi.left_fit_coeff, 
                                                                                                vfi.right_fit_coeff)
            self.push_lane_info(vfi)
            
        # if the queue is full, quick search is done using polynomial of prvious frame
        # if the lane detection does not make sense, the frames of failed detection are skippied
        # until a failure tolerance is reached.
        # Once the tolrence is reached, the info queue is deleted and sliding-window search is done again.
        
        else:
            last_frame = self.lane_info_queue[len(self.lane_info_queue)-1]
            vfi.leftx, vfi.lefty, vfi.rightx, vfi.righty = self.search_around_poly(warped_frame, 
                                                                                   last_frame.left_fit_coeff, 
                                                                                   last_frame.right_fit_coeff)
            vfi.left_fit_coeff, vfi.right_fit_coeff = self.fit_polynomial(vfi.leftx, vfi.lefty, vfi.rightx, vfi.righty)
            vfi.left_real_curvature, vfi.right_real_curvature = self.measure_curvature_real(vfi.left_fit_coeff, 
                                                                                            vfi.right_fit_coeff)
            vfi.car_offset_distance, vfi.car_offset_side, vfi.center_of_lane = self.find_offset(vfi.left_fit_coeff, 
                                                                                                vfi.right_fit_coeff)
            
            if self.check_sanity(vfi):
                self.push_lane_info(vfi)
                
            elif self.failed_frames_count < self.failure_tolerance:
                self.failed_frames_count = self.failed_frames_count +1
                vfi = last_frame          # Over write the detected values, with that of last-frame
            
            else:
                self.failed_frames_count = 0              # Reset the count
                self.lane_info_queue = []                 # Reset the queue
                vfi = last_frame          # Over write the detected values, with that of last-frame
                #return self.process_frame(frame_image)
            
        annotated_frame = self.plot_on_real_frame(undistorted_frame, warped_frame, vfi.left_fit_coeff, vfi.right_fit_coeff)
        cv2.putText(annotated_frame,
                    "Radius of Curvature :"+str((vfi.left_real_curvature + vfi.right_real_curvature)/2)[:5]+"(m)",
                    (100,100), cv2.FONT_HERSHEY_TRIPLEX, 1, [255,255,255], 2)
        #print ("Average Radius of curvature :", str((vfi.left_real_curvature + vfi.right_real_curvature)/2))
        cv2.putText(annotated_frame,
                    "The car is "+str(vfi.car_offset_distance)[:3]+" meters, "+vfi.car_offset_side+" of centre of lane",
                    (100,150), cv2.FONT_HERSHEY_TRIPLEX, 1, [255,255,255], 2)
        
        return annotated_frame            
    
    
```
